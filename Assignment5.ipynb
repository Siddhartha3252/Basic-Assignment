```python
1. What does an empty dictionary's code look like?

Answer :
    
    Empty dictionary's code look like pair curly brackets: {}
```

```python
2. What is the value of a dictionary value with the key 'foo' and the value 42?

Answer :
    
    The value of a dictionary value with the key 'foo' and the value 42 is {'foo': 42}.
```

```python
3. What is the most significant distinction between a dictionary and a list?

Answer :
    
    Items stored in a dictionary are unordered, where as the items in a list are ordered.
```

```python
6. If a dictionary is stored in spam, what is the difference between the expressions 'cat' in spam and 'cat' in spam.values()?

Answer :
    
    'cat' in spam checks whether there is a 'cat' key in the dictionary, where as 'cat' in spam.values() checks if there is a value 'cat' for one of the keys in spam.
```
